### Identify the top 10 handsets used by the customers.

```python
# Assuming 'Handset Type' is the column representing the model of the handsets
df_filter = df[df['Handset Type'] != 'undefined']

top_10_handsets = df_filter['Handset Type'].value_counts().head(10)

# Display the top 10 handsets
print(f'Top 10 Handsets: {top_10_handsets}')
```

### Identify the Top 3 Handset Manufacturers

```python
# Find the top 3 handset manufacturers
top_3_manufacturers = df['Handset Manufacturer'].value_counts().head(3)

# Display the top 3 manufacturers
print(top_3_manufacturers)
```

### Identify the Top 5 Handsets Per Top 3 Manufacturer

```python
# Loop through the top 3 manufacturers to find the top 5 handsets for each
for manufacturer in top_3_manufacturers.index:
    print(f"Top 5 handsets for {manufacturer}:")
    top_5_handsets = df[df['Handset Manufacturer'] == manufacturer]['Handset Type'].value_counts().head(5)
    print(top_5_handsets)
    # print("\n")
```
